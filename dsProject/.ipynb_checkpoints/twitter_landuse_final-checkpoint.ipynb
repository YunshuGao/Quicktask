```python
import pandas as pd
```

```python
import geopandas
```

```python
zipfile=r"D:\RA-Tasks\1 Graph neural network_POI\Task1\Process\twitterPoint_final.shp"
df=geopandas.read_file(zipfile)
df.head(5)
```

```python
df1=df[['OBJECTID_1','Field1','created_at','id_str','lang','time_zone','LanduseTyp','geometry']]
```

```python
df2=pd.read_csv(r"D:\RA-Tasks\1 Graph neural network_POI\Task1\Process\result_v2.csv")
```

```python
df2.rename(columns={ df2.columns[0]: "Field1" }, inplace = True)
```

```python
df1.head(5)
```

```python
df2.head(5)
```

```python
df3 = pd.merge(df1, df2, on='Field1', how='inner')
```

```python
df3.head(5)
```

```python
df4=df3[['OBJECTID_1','Field1','LanduseTyp','created_at_x','id_str_x','time_zone_x','geometry','date_thirdprocess','date_hk']]
```

```python
df4['date_thirdprocess']= pd.to_datetime(df4['date_thirdprocess'])
df4['date_hk']= pd.to_datetime(df4['date_hk'])
```

```python
date_hk_col=pd.DatetimeIndex(df4['date_hk'])
```

```python
import matplotlib.pyplot as plt
df4['id_str_x'].groupby(df4['LanduseTyp']).count()
df4['id_str_x'].groupby(df4['LanduseTyp']).count().plot(kind="bar")
```

```python
# by year
df4['id_str_x'].groupby([date_hk_col.year,df4['LanduseTyp']]).count()
```

```python
# by month
df4['id_str_x'].groupby([date_hk_col.month,df4['LanduseTyp']]).count()
```

```python
# by day
df4['id_str_x'].groupby([date_hk_col.day,df4['LanduseTyp']]).count()
```

```python
# by hour
df4['id_str_x'].groupby([date_hk_col.hour,df4['LanduseTyp']]).count()
```
